# Creating Torch Dataset Class

## Pytorch's Abstract Dataset Class

```python
class Dataset(object):
    
    def __getitem__(self, index):
        raise NotImplementedError
    
    def __len__(self):
        raise NotImplementedError
        
    def __add__(self, other):
        return ConcatDataset([self, other])
```

## Creating Custom MRI Class

```python
class MRI(Dataset):
    def __init__(self):
        
        # READing All images
        tumor = []
        healthy = []
        path_yes = "./brain_tumor_dataset/yes/*.jpg"
        for file in glob.iglob(path_yes):
            img = cv2.imread(file)
            img = cv2.resize(img, (128, 128))
            b, g, r = cv2.split(img)
            img = cv2.merge((r,g,b))
            img = img.reshape((img.shape[2], img.shape[0], img.shape[1]))
            #print(img.shape)
            tumor.append(img)
        path_no = "./brain_tumor_dataset/no/*.jpg"
        for file in glob.iglob(path_no):
            img = cv2.imread(file)
            img = cv2.resize(img, (128, 128))
            b, g, r = cv2.split(img)
            img = cv2.merge((r,g,b))
            img = img.reshape((img.shape[2], img.shape[0], img.shape[1]))
            #print(img.shape)
            healthy.append(img)
        
        # Our images:
        tumor = np.array(tumor, dtype=np.float32)
        healthy = np.array(healthy, dtype=np.float32)
        #print(healthy.shape)
        
        
        # Our labels:
        tumor_label = np.ones(tumor.shape[0], dtype=np.float32)
        healthy_label = np.zeros(healthy.shape[0], dtype=np.float32)
        
        # Concatenating tumor and healthy images and lables
        self.images = np.concatenate((tumor, healthy), axis = 0)
        #print(self.images.shape)
        self.labels = np.concatenate((tumor_label, healthy_label))
        #print(self.labels)
    
    def __getitem__(self, index):
        sample = {"image": self.images[index], "label": self.labels[index]}
        return sample
    
    def __len__(self):
        return self.images.shape[0]
    
    def normalize(self):
        self.images = self.images/255
```

```python
mri = MRI()
mri.normalize()
```

# Creating a CNN model

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()
        self.cnn_model = nn.Sequential(
        nn.Conv2d(in_channels = 3, out_channels = 6, kernel_size = 5),
        nn.Tanh(),
        nn.AvgPool2d(kernel_size = 2, stride = 5),
        nn.Conv2d(in_channels = 6, out_channels = 16, kernel_size = 5),
        nn.Tanh(),
        nn.AvgPool2d(kernel_size = 2, stride = 5)        
        )
        
        self.fc_model = nn.Sequential(
        nn.Linear(in_features = 256, out_features = 120),
        nn.Tanh(),
        nn.Linear(in_features = 120, out_features = 84),
        nn.Tanh(),
        nn.Linear(in_features = 84, out_features = 1),
        )
        
    def forward(self, x):
        x = self.cnn_model(x)
        x = x.view(x.size(0), -1) # This line will faltten the 2D arrays
        x = self.fc_model(x)
        x = torch.sigmoid(x)
        return x
```

# Evaluating the first model

```python
mri_dataset = MRI()
mri_dataset.normalize()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = CNN().to(device)
```

```python
dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=False)
```

```python
model.eval()
outputs = []
y_true = []
with torch.no_grad():
    for D in dataloader:
        image = D['image'].to(device)
        label = D['label'].to(device)

        y_hat = model(image)

        outputs.append(y_hat.cpu().detach().numpy())
        y_true.append(label.cpu().detach().numpy())
    
```

```python
outputs = np.concatenate( outputs, axis=0 ).squeeze()
y_true = np.concatenate( y_true, axis=0 ).squeeze()
```

```python
def threshold(scores,threshold=0.50, minimum=0, maximum = 1.0):
    x = np.array(list(scores))
    x[x >= threshold] = maximum
    x[x < threshold] = minimum
    return x
```

# Training the first model

```python
eta = 0.0001
EPOCH = 400
optimizer = torch.optim.Adam(model.parameters(), lr=eta)
dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=True)
model.train()
```

```python
for epoch in range(1, EPOCH):
    losses = []
    for D in dataloader:
        optimizer.zero_grad()
        data = D['image'].to(device)
        label = D['label'].to(device)
        y_hat = model(data)
        # define loss function
        error = nn.BCELoss() 
        loss = torch.sum(error(y_hat.squeeze(), label))
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
    if (epoch+1) % 10 == 0:
        print('Train Epoch: {}\tLoss: {:.6f}'.format(epoch+1, np.mean(losses)))
```

# Evaluating the new Model

```python
model.eval()
dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=False)
outputs=[]
y_true = []
with torch.no_grad():
    for D in dataloader:
        image =  D['image'].to(device)
        label = D['label'].to(device)
        y_hat = model(image)
        
        outputs.append(y_hat.cpu().detach().numpy())
        y_true.append(label.cpu().detach().numpy())
        
        
outputs = np.concatenate( outputs, axis=0 )
y_true = np.concatenate( y_true, axis=0 )
```
